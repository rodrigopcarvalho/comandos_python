```python
import pandas as pd
```

```python
# Lendo arquivos em Excel.

df1 = pd.read_excel('/content/Aracaju.xlsx')
df2 = pd.read_excel('/content/Fortaleza.xlsx')
df3 = pd.read_excel('/content/Natal.xlsx')
df4 = pd.read_excel('/content/Recife.xlsx')
df5 = pd.read_excel('/content/Salvador.xlsx')
```

```python
# Juntando todos os Arquivos>

df = pd.concat([df1, df2, df3, df4, df5])
```

```python
df.head()
```

```python
df.tail()
```

```python
# Pegando uma Amostra Aleatoria de Dados.

df.sample(5)
```

```python
# Tipos de Dados.

df.dtypes
```

```python
# Alterando o tipo de dados de uma coluna. de Inteiro pa String.

df['LojaID'] = df['LojaID'].astype('object')
```

```python
df.dtypes
```

```python
# Consultando Linhas com valores faltantes.

df.isnull().sum()
```

```python
# Substituindo os valores nulos pela média.

df['Vendas'].fillna(df['Vendas'].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
# Substituindo os valores nulos por 0.

df['Vendas'].fillna(0, inplace = True)
```

```python
# Apagandos a linhas com os valores nulos.

df.dropna(inplace = True)
```

```python
# Apagando os valores nulos de apenas uma coluna.

df.dropna(subset=['Vendas'], inplace = True)
```

```python
# Removendo linhas que estejam com valores faltantes em todas as colunas.

df.dropna(how = 'all', inplace = True)
```

```python
# Criando novas colunas.

df['Receitas'] = df['Vendas'].mul(df['Qtde'])
```

```python
df.head()
```

```python
# Retornando a maior receita.

df['Receitas'].max()
```

```python
df['Receitas'].min()
```

```python
# Método nlargest()
# Retorna os maiores dados das colunas solicitas.

df.nlargest(3, 'Receitas')
```

```python
# Método nsmallest()
# Retorna os menores dados das colunas solicitas.

df.nsmallest(3, 'Receitas')
```

```python
# Agrupando dados.

df.groupby('Cidade')['Receitas'].sum()
```

```python
# Ordenando os conjuntos de Dados. do maior para o menor.

df.sort_values('Receitas', ascending = False).head(10)
```

# Trabalhando com Datas

```python
# Transformando a coluna de Data em tipo inteiro

df['Data'] = df['Data'].astype('int64')
```

```python
# Verificando o tipo de cada coluna.

df.dtypes
```

```python
# Transformando a coluna de data em data.

df['Data'] = pd.to_datetime(df['Data'])
```

```python
df.dtypes
```

```python
# Agrupamento por ano.

df.groupby(df['Data'].dt.year)['Receitas'].sum()
```

```python
# Criando uma coluna com o ano.

df['Ano_Venda'] = df['Data'].dt.year
```

```python
df.sample(5)      
```

```python
# Extraindo o mês e o dia.

df['mes_venda'], df['dia_venda'] = (df['Data'].dt.month, df['Data'].dt.day)
```

```python
df.sample(5)
```

```python
# Retornando a data mais antiga.

df['Data'].min()
```

```python
# Criando uma nova coluna com a diferença de dias.

df['diferenca_dias'] = df['Data'] - df['Data'].min()
```

```python
df.sample(5)
```

```python
# Criando a Coluna de trimestre.

df['trimestre_vendas'] = df['Data'].dt.quarter
```

```python
df.sample(5)
```

```python
# filtrando as vendas de 2019 do mês de Março.

vendas_marco_19 = df.loc[(df['Data'].dt.year == 2019) & (df['Data'].dt.month == 3)]
```

```python
vendas_marco_19
```

```python
vendas_marco_19.sample(10)
```
